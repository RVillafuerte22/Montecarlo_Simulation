### Definici√≥n del problema:

Cada a√±o, en los meses de octubre las Empresas del sector financiero, tales como Banco, 
Financieras y Cajas Municipales requieren proyectar sus ingresos y gastos correspondientes al 
a√±o pr√≥ximo, en este caso para el a√±o 2022. Con la aplicaci√≥n de simulaci√≥n Montecarlo se 
pretende ayudar, espec√≠ficamente a Financiera Confianza, en aproximar como terminar√°n sus
ingresos y gastos proyectados.
La fuerza de ventas considera asesores comerciales, quienes tienen como responsabilidad 
contactar a los clientes, evaluarlos y si califican procesar y desembolsar el cr√©dito.
Los ingresos resultan de la multiplicaci√≥n del saldo de cartera y de la tasa de inter√©s a la cual
se coloca el cr√©dito. El saldo de cartera es resultado de la f√≥rmula que sigue:

ùë†ùëéùëôùëëùëú ùëêùëéùëüùë°ùëíùëüùëé ùëìùëñùëõùëéùëô = ùë†ùëéùëôùëëùëú ùëêùëéùëüùë°ùëíùëüùëé ùëñùëõùëñùëêùëñùëéùëô + ùëëùëíùë†ùëíùëöùëèùëúùëôùë†ùëúùë† ‚àí ùëéùëöùëúùëüùë°ùëñùëßùëéùëêùëñ√≥ùëõ

Donde saldo cartera inicial es el saldo al cierre de mes anterior, desembolsos es el monto 
colocado en el mes en curso y la amortizaci√≥n los pagos al capital de los cr√©ditos.
Para el monto de desembolso se requiere considerar lo siguiente:

ùê∑ùëíùë†ùëíùëöùëèùëúùëôùë†ùëúùë† = ùëáùëñùëêùëòùëíùë° ùë• ùëÉùëüùëúùëëùë¢ùëêùë°ùëñùë£ùëñùëëùëéùëë ùë• ùëÅùëüùëú ùëéùë†ùëíùë†ùëúùëüùëíùë†

Donde el ticket es el monto promedio de cr√©dito otorgado al cliente, productividad es el 
n√∫mero de operaciones que realiza el asesor comercial en el mes y n√∫mero de asesores es la 
cantidad de asesores comerciales (vendedores de servicios financieros) que dispone la 
Financiera en un mes dado.
La amortizaci√≥n es una funci√≥n del saldo de cartera del mes anterior. De la data 
proporcionada por la Financiera, se tiene que la amortizaci√≥n es un 10% del saldo de cartera del 
mes anterior. Esta aproximaci√≥n se usar√° al momento de realizar la proyecci√≥n.

### 2. OBJETIVO:

Estimar el margen de ganancia/p√©rdida de la Financiera para el a√±o 2022 utilizando t√©cnicas 
de simulaci√≥n Montecarlo.

### 3.VARIABLES ASOCIADAS AL ESTUDIO

Saldo de cartera: Stock de cartera de activas a una determinada fecha de corte

Ticket: Promedio de monto desembolso del mes

Productividad: N√∫mero de operaciones realizadas por el asesor en un mes

Nro de asesores: Cantidad de asesores comerciales (vendedores) 

Amortizaci√≥n: Pagos al capital por los pr√©stamos de dinero realizados.

Tasa de inter√©s: Tasa promedio ponderada a la cu√°l se desembolsan los cr√©ditos

Gastos financieros: Pago de interes a los ahorristas, pago de fondeo

Gastos operativos: Sueldos de colaboradores

Gastos generales: Pago de servicios de agua, luz, internet, alquiler de local, etc.

Gastos de provisi√≥n: Gasto por deterioro de cartera
